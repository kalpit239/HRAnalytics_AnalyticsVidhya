```python
#CodeSection10
# Write the Transformation Code
numerical = ['city_development_index', 'training_hours']
for num_var in numerical:
  for dataset in all_data:
    dataset[num_var] = np.log(dataset[num_var]+1)
 # "1" is added to avoid error for rows that contain "0"
```

## 6/ Impute Categorical Variables
- Lets try another Strategy of Imputing "Missing" where a Category value is missing

### Impute all Categorical Variables at once

```python
#CodeSection14
categorical = ['gender','relevent_experience','enrolled_university','education_level','major_discipline','experience','company_size','company_type', 'last_new_job']
for cat_var in categorical:
  for dataset in all_data:
    dataset[cat_var] = dataset[cat_var].fillna('Missing')
```

```python
#CodeSection17
# Get X and y Variables for your Dataset
X = train.drop(['enrollee_id','target'], axis =  1)
y = train['target']
```

### Create Dummy Variables for  Categorical Variables

```python
#CodeSection18
# Get Dummy Variables
X = pd.get_dummies(X)  # Keep only n - 1 Variables - Use "drop_first = True"
```

## 9/ Build Model

```python
#CodeSection19
#Import Logistic Regression Library
from sklearn.linear_model import LogisticRegression

#Initialize LR
clf = LogisticRegression(max_iter = 1000, class_weight = 'balanced')
```

```python
#CodeSection20
# Fit Model
model = clf.fit(X, y)
```

### Check Accuracy of Model on Train Data

```python
#CodeSection21
# Predict on Train Data
y_pred = model.predict(X)
```

### We can use a Accuracy Function from Metrics
- Check Train Accuracy

```python
#CodeSection22
# Import metrics library
from sklearn.metrics import accuracy_score
```

```python
#CodeSection23
# get Actual "y" variables (use "y_true" as variable)
y_true = y
```

```python
#CodeSection24
# Print Train Accuracy
print(f" Train Accuracy : {accuracy_score(y_true, y_pred):0.1%}")
```
